```python
import pandas as pd

df= pd.read_csv("adult-all-1.csv", header=None)
df.head()
```

```python
df.columns = [
    "age", "workclass", "fnlwgt", "education", "education_num",
    "marital_status", "occupation", "relationship", "race", "sex",
    "capital_gain", "capital_loss", "hours_per_week", "native_country", "income"
]

df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.replace(" ?", pd.NA)
df = df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="income", data=df)
plt.title("Income Distribution")
plt.xlabel("Income Level")
plt.ylabel("Count")
plt.show()
```

This chart shows how income is distrubted across the dataset. Alot of the individuals fall into the income category of 50k or less while a small percentage earns above 50k. This shows how uneven the income distribution is and shows that higher income levels are less common based on the overall population

```python
plt.figure(figsize=(10,6))
sns.countplot(y="education", hue="income", data=df)
plt.title("Education Level vs Income")
plt.xlabel("Count")
plt.ylabel("Education Level")
plt.legend(title="Income")
plt.show()
```

This visual is comparing education level with income. People with a higher level of education like having a bachelors degree or higher are more than likely to earn above 50k. Individuals with a lower education level are mostly category of 50k or less. This shows that education plays a big role when it comes to earnings long term

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="income", y="age", data=df)
plt.title("Age Distribution by Income Level")
plt.xlabel("Income Level")
plt.ylabel("Age")
plt.show()
```

This box plot shows the relationship between age and income. People who earn above 50k are older compared to those earning 50k or less. This pattern shows that income may increase as you age due to gaining more work experience and improving on skills.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="income", y="hours_per_week", data=df)
plt.title("Hours Worked per Week vs Income")
plt.xlabel("Income Level")
plt.ylabel("Hours per Week")
plt.show()
```

This chart looks at how many people work per week based on income level. People who earn above 50k work more hours per week compared to those who are in the 50k or less lower income group. This shows that higher income roles are in need for longer working hours or more job responsibility.

```python
plt.figure(figsize=(10,6))
sns.countplot(y="occupation", hue="income", data=df)
plt.title("Occupation vs Income Level")
plt.xlabel("Count")
plt.ylabel("Occupation")
plt.legend(title="Income")
plt.show()
```

This visual compares income levels in different job posiitons. White collared jobs have a high percentage of people earning above 50k while blue collared jobs are shown to make 50k or less. Depending on what job you have has a huge influence on how much income you get.

This analysis shows that income is affected based on your education, age, occupation, and hours you work per week. People with higher education and white collared jobs are more likely to earn above 50k while those those who are less educated and have blue collared jobs tend to fall in the lower income group. These patters show how much socio economic factors can affect every persons source of income
